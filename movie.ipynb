```python
import duckdb
import pandas as pd

# Sample movie data
data = {
    'Title': ['Movie A', 'Movie B', 'Movie C', 'Movie D'],
    'Genre': ['Comedy', 'Action', 'Comedy', 'Drama'],
    'Release Date': ['2001-01-01', '2010-05-20', '2015-07-10', '2020-12-15'],
    'Revenue': [150000000, 350000000, 400000000, 100000000]
}

# Create DataFrame
df_movies = pd.DataFrame(data)

# Register the DataFrame with DuckDB
duckdb.register('movies', df_movies)

# Now run your SQL queries using duckdb.query().df() to see results as a DataFrame
```

```python
duckdb.query("""
SELECT * FROM movies
ORDER BY "Release Date"
""").df()
```

```python
duckdb.query("""
SELECT * FROM movies
ORDER BY "Release Date" DESC
""").df()
```

```python
duckdb.query("""
SELECT * FROM movies
WHERE Genre = 'Comedy'
ORDER BY "Release Date" DESC
""").df()
```

```python
duckdb.query("""
    SELECT *
    FROM read_csv_auto('Movie-Data.csv')
    WHERE Genre = 'Comedy'
    ORDER BY "Release Date" DESC
""").df()
```

```python
duckdb.query("""
SELECT * FROM movies
WHERE Genre = 'Comedy'
AND Revenue > 300000000
ORDER BY "Release Date" DESC
""").df()
```
