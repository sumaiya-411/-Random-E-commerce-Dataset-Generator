```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")


data = {
    'Name': ['junaid', 'sumaiya', 'jawad', 'adil', 'juveria'],
    'Age': [25, 19, 34, 40, 22],
    'Department': ['data scientist', 'Engineering', 'Marketing', 'Finance', 'data scientist'],
    'Salary': [80000, 60000, 55000, 65000, 35000]
}

df = pd.DataFrame(data)


print("First 5 rows of the dataset:")
print(df.head())



print("\nNumber of Duplicate Rows:")
print(df.duplicated().sum())


print("\nNumber of People in Each Department:")
print(df['Department'].value_counts())


print("\nAverage Salary of All Employees:")
print(df['Salary'].mean())




plt.figure(figsize=(6, 4))
sns.countplot(x='Department', data=df)
plt.title('Count of People in Each Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.tight_layout()
plt.show()


plt.figure(figsize=(6, 4))
sns.boxplot(x='Salary', data=df)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.tight_layout()
plt.show()


plt.figure(figsize=(6, 4))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```
