```python
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'age': [25, 30, 35, 40, np.nan],
    'gender': ['F', 'M', 'M', 'M', 'F'],
    'income': [50000, 60000, 70000, 80000, 75000]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Fill missing age with average
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
# Average income by gender
income_by_gender = df.groupby('gender')['income'].mean()
print(income_by_gender)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart of income by person
plt.figure(figsize=(8, 4))
sns.barplot(x='name', y='income', data=df)
plt.title('Income by Person')
plt.show()

# Boxplot of income by gender
sns.boxplot(x='gender', y='income', data=df)
plt.title('Income Distribution by Gender')
plt.show()
```
