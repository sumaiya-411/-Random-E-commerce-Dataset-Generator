```python
from textblob import TextBlob
import matplotlib.pyplot as plt
from collections import Counter

# Sample Data from Different Sources
amazon_reviews = [
    "This product is amazing, I love it!",
    "Terrible quality, broke after one use.",
    "The product is okay for the price."
]

social_media_posts = [
    "I had a great experience shopping online today!",
    "I'm so frustrated with the customer service.",
    "Anyone else thinks this app is just...meh?"
]

news_articles = [
    "The company's new launch is receiving positive feedback from the public.",
    "There is growing concern over the product safety issues.",
    "Mixed reactions as the company expands its services globally."
]

# Combine all reviews
all_texts = amazon_reviews + social_media_posts + news_articles

# Analyze Sentiment
sentiments = []

for text in all_texts:
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity

    if polarity > 0:
        sentiments.append("Positive")
    elif polarity < 0:
        sentiments.append("Negative")
    else:
        sentiments.append("Neutral")

# Show Results
print("Sentiment Analysis Results:\n")
for idx, text in enumerate(all_texts):
    print(f"Source {idx+1}: {text}")
    print(f"Sentiment: {sentiments[idx]}\n")

# Count the sentiments
sentiment_counts = Counter(sentiments)

# Plotting
plt.figure(figsize=(6,6))
plt.pie(sentiment_counts.values(), labels=sentiment_counts.keys(), autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'lightcoral', 'lightblue'])
plt.title('Public Sentiment Analysis')
plt.show()
```
