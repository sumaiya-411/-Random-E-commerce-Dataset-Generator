```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Create an enhanced sample dataset
data = {
    'Name': ['Syed', 'Henry', 'Noor', 'Matt', 'Micky', 'Alfred', 'Stephen', 'Campbell', 'Mannan', 'Masimo'],
    'Age': [20, 45, 24, 39, 72, 23, 8, 16, 21, 7],
    'Score': [85, 72, 90, 65, 88, 78, 92, 68, 75, 95],
    'Gender': ['M', 'M', 'F', 'M', 'M', 'M', 'M', 'M', 'M', 'M']  
}

df = pd.DataFrame(data)

# Step 3: Show the dataset with basic statistics
print("Sample Data:")
print(df)
print("\nBasic Statistics:")
print(df.describe())

# Step 4: Bar Plot - showing actual ages instead of counts
plt.figure(figsize=(10, 6))
barplot = sns.barplot(data=df, x='Name', y='Age', palette='viridis')
plt.title("Age Distribution by Name", fontsize=14, pad=20)
plt.xlabel("Name", fontsize=12)
plt.ylabel("Age", fontsize=12)
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=45, ha='right')
plt.tight_layout()
plt.savefig("age_by_name.png", dpi=300, bbox_inches='tight')
plt.show()

# Step 5: Enhanced Histogram with density curve
plt.figure(figsize=(8, 5))
hist = sns.histplot(data=df, x='Age', bins=8, kde=True, 
                   color='skyblue', edgecolor='navy', linewidth=1)
plt.title("Age Distribution with Density Curve", fontsize=14, pad=15)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.savefig("age_distribution.png", dpi=300, bbox_inches='tight')
plt.show()

# Step 6: Enhanced Correlation Heatmap with proper variables
plt.figure(figsize=(6, 4))
corr_matrix = df[['Age', 'Score']].corr()  
heatmap = sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', 
                     linewidths=0.5, linecolor='white', 
                     vmin=-1, vmax=1, fmt=".2f",
                     annot_kws={"size": 12})
plt.title("Age vs Score Correlation", fontsize=14, pad=15)
plt.tight_layout()
plt.savefig("correlation_heatmap.png", dpi=300, bbox_inches='tight')
plt.show()

# Step 7: Score by Gender
plt.figure(figsize=(6, 5))
boxplot = sns.boxplot(data=df, x='Gender', y='Score', palette='pastel')
plt.title("Score Distribution by Gender", fontsize=14, pad=15)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.savefig("score_by_gender.png", dpi=300, bbox_inches='tight')
plt.show()
```
